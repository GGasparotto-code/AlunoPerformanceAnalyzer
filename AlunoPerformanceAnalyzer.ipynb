Acessando o arquivo .csv utilizando somente python

```python
# Importando o módulo csv para trabalhar com arquivos CSV
import csv

# Abrindo o arquivo alunos.csv para leitura
f = open("alunos.csv", "r", encoding="utf-8")

# Usando o objeto csv.reader para ler o conteúdo do arquivo
reader = csv.reader(f, delimiter=';', lineterminator='\n')

# Inicializando uma lista vazia para armazenar as linhas lidas do arquivo
worksheet = []

# Iterando sobre cada linha do arquivo
for row in reader:

  # Adicionando cada linha à lista worksheet
  worksheet.append(row)

# Fechando o arquivo após a leitura
f.close()

# Imprimindo o conteúdo da lista worksheet
print(worksheet)
```

```python
# Utilizando uma compreensão de lista para extrair a segunda coluna (índice 1) de cada linha em worksheet
[item[1] for item in worksheet]
```

```python
# Importando a biblioteca pandas para manipulação de dados
import pandas as pd
```

```python
# Lendo o arquivo alunos.csv usando pandas, especificando o delimitador como ';'
df = pd.read_csv("alunos.csv", sep=';')
df
```

```python
# Mostrando os nomes das colunas do DataFrame
df.columns
```

```python
# Acessando a terceira linha (índice 2) do DataFrame
df.loc[2]
```

```python
# Calculando a média das notas de quatro provas para criar uma nova coluna chamada "Media"
df["Media"] = df[["Prova_1", "Prova_2", "Prova_3", "Prova_4"]].mean(axis=1)
df
```

```python
# Informações detalhadas sobre o DataFrame
df.info()
```

```python
# Salvando o DataFrame em um novo arquivo CSV, sem incluir o índice
df.to_csv("medias_processadas.csv", sep=';', index=False)
```

```python
# Salvando o DataFrame em um arquivo Excel, também sem incluir o índice
df.to_excel("medias_processadas.xlsx", index=False)
```
